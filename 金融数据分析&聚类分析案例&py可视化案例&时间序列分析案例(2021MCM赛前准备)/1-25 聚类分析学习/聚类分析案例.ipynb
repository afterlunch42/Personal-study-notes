```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
%matplotlib inline
```

##  读取数据

```python
data = pd.read_excel('./consumption_data.xls', sheet_name='Sheet1')
data.head()
```

## 数据可视化

```python
data[['R', 'F', 'M']].plot(cmap='brg', subplots=True)
```

## 数据预处理

```python
data = data.dropna()  # 去除空值
s_data = (data[['R', 'F','M']]-data[['R', 'F','M']].mean(axis=0))/(data[['R', 'F','M']].std(axis=0))  # 数据标准化(z-score)
data.head()
s_data['id'] = data['Id']
s_data
```

```python
kmodel = KMeans(n_clusters=10, n_jobs=8)  # n_jobs是并行数，一般等于CPU数较好
kmodel.fit(s_data[['R', 'F', 'M']])  ##训练模型 
```

```python
label = pd.Series(kmodel.labels_)  # 各样本的类别
num = pd.Series(kmodel.labels_).value_counts()  # 统计各样本对应的类别的数目
center = pd.DataFrame(kmodel.cluster_centers_)  # 找出聚类中心 
center_max = center.values.max()
center_min = center.values.min()
```

```python
from mpl_toolkits.mplot3d import Axes3D  # 用于绘制3D散点图
```

```python
result = data
result['c_r'] = kmodel.labels_  # 将聚类类别序号写入数据中

# plt.style.use('seaborn')
plt.style.use('ggplot')  # plt使用色彩风格
fig = plt.figure()  # 生成plt对象
c_list= ['#6A5ACD','#B0E0E6','#778899', '#00FF7F','#228B22', '#F0E68C'
        ,'#FF8C00','#FA8072','#6495ED','#BA55D3']  # 色值列表
ax = fig.add_subplot(111, projection='3d')  # 创建3D图像底版
for i in range(0, 10):  # 提取每一类别的数据并绘图
    c_data = result[result['c_r']==i]
    x0 = c_data['R'].values  # 转为array
    y0 = c_data['F'].values
    z0 = c_data['M'].values
    color = c_list[i]  # 选择颜色
    ax.scatter(x0, y0, z0, c=color, marker='.')  # 绘制3D散点图
ax.set_xlabel('R')
ax.set_ylabel('F')
ax.set_zlabel('M')
plt.title('KMeans result')
fig = plt.gcf()  # 捕捉当前图像信息，用于保存图片
plt.show()

fig.savefig('clustering.png', dpi=800)  # 保存图片
```
